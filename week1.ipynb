### 1. GCD

```python
# Method 1 - store the gcd
def gcd1(a, b):
    cf = [] # store the common factors
    for i in range(1, int(min(a, b)/2) + 1):     
        # we should go till half of the min. (inc performance to double)
        if a % i == 0 and b % i == 0:
            cf.append(i) 
    return(cf[-1])  # return the last one

gcd1(88874608626, 368705866)
```

```python
# Method 2 - Keep a flag
def gcd2(a, b):
    gcd = 1
    cf = [] # store the common factors
    for i in range(1, int(min(a, b)/2) + 1):     
        # we should go till half of the min. (inc performance to double)
        if a % i == 0 and b % i == 0:
            gcd = i
    return(gcd)  # Here the space is less required

gcd2(88874608626, 368705866)
```

### 2. Counting Primes

```python
# Q1 - Primes upto m

import math

def is_prime(n):
    if n == 1: return False
    for i in range(2, int(math.sqrt(n))+1):
        if n%i == 0:
            return False
    return True

def is_prime2(n):
    if n == 1: return False
    (result, i) = (True, 2)
    while(result and i <= math.sqrt(n)):
        if n%i == 0:
            result = False
        i += 1
    return(result)


def primesupto(m):
    pl = []     # stores the primes
    for i in range(1, m+1):
        if is_prime(i):
            pl.append(i)
    return(len(pl))

primesupto(600000)
```

```python
# Q2 - First m primes
def firstprimes(m):
    (count, i, pl) = (0, 1, [])
    while(count < m):
        if is_prime(i):
            pl.append(i)
            count += 1
        i += 1
    return(pl)
firstprimes(10)
```

#### Difference between two cons. primes - count

```python
def primediffs(n):
    lastprime = 2
    pd = {}     # Stores the prime differences
    for i in range(3, n+1):
        if is_prime(i):
            diff = i - lastprime
            lastprime = i

            if diff in pd.keys():
                pd[diff] += 1
            else:
                pd[diff] = 1

    return pd

primediffs(1000)
```
